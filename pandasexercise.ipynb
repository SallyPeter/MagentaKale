```python
# c. Analyze crime trends over time by grouping data by month

month_crime = pd.DataFrame(data.groupby(by=["Month", "Crime type"]).count()["Crime ID"])
month_crime
```

```python
# The highest recorded crime per month

for month in month_crime.Month.unique():
    month_data = month_crime[month_crime.Month == month]
    print(month_data[month_data["Crime ID"] == month_data["Crime ID"].max()])
```
