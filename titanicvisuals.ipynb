```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Set the style
sns.set_style("whitegrid")
```

```python
filepath = r"C:\Users\esltpae\Downloads\titanic"
data = pd.read_csv(filepath + "\\" + "train.csv")
data.head()
```

```python
survived = data.Survived.value_counts()
plt.bar(survived.index, survived.values);

plt.title("Count of those who survived or not");
plt.ylabel("Count");
plt.xlabel("Survived");
plt.xticks(ticks=[0,1], labels=["Died", "Survived"]);
```

```python
sns.countplot(x="Survived", data=data, palette="coolwarm");

plt.title("Count of those who survived or not");
plt.ylabel("Count");
plt.xlabel("Survived");
plt.xticks(ticks=[0,1], labels=["Died", "Survived"]);
```

```python
sex = data.Sex.value_counts()
plt.bar(sex.index, sex.values);

plt.title("Gender Distribution");
plt.ylabel("Count");
plt.xlabel("Sex");
```

```python
sns.countplot(x="Sex", data=data, palette="pastel");

plt.title("Gender Distribution");
plt.ylabel("Count");
plt.xlabel("Sex");
```

```python
pclass = data.Pclass.value_counts()
plt.bar(pclass.index, pclass.values);

plt.title("Pclass Distribution");
plt.ylabel("Count");
plt.xlabel("Pclass");
plt.xticks(ticks=[1,2,3], labels=['1','2','3']);
```

```python
sns.countplot(x="Pclass", data=data, palette="pastel");

plt.title("Pclass Distribution");
plt.ylabel("Count");
plt.xlabel("Pclass");
```

```python
plt.hist(data.Age, bins=30);
plt.title("Age Distribution");
plt.ylabel("Count");
plt.xlabel("Age");
```

```python
sns.histplot(data["Age"].dropna(), bins=30, kde=True, color="blue")
plt.title("Age Distribution");
plt.ylabel("Count");
plt.xlabel("Age")
```

```python
surclass = data.groupby(["Pclass", "Survived"])["PassengerId"].count().unstack()#.reset_index()
surclass.plot(kind='barh', stacked=False)
```

```python
surclass = data.groupby(["Pclass", "Survived"])["PassengerId"].count().unstack()#.reset_index()
surclass.plot(kind='barh', stacked=True);
plt.title("Survival Rate by Class");
```

```python
sns.barplot(x="Pclass", y="Survived", data=data, palette="magma");
plt.title("Survival Rate by Class");
```

```python
sns.barplot(x="Sex", y="Survived", data=data, palette="coolwarm")
plt.title("Survival Rate by Gender")
```

```python
surclass = data.groupby(["Sex", "Survived"])["PassengerId"].count().unstack()#.reset_index()
surclass.plot(kind='barh', stacked=True);
plt.title("Survival Rate by Sex");
```

```python
plt.figure(figsize=(8, 4))
sns.boxplot(x='Pclass', y='Fare', data=data, palette='coolwarm')
plt.title("Fare Distribution by Class")
```

```python
plt.figure(figsize=(8, 4))
sns.boxplot(x='Pclass', y='Fare', data=data, palette='coolwarm')
plt.title("Fare Distribution by Class")
plt.ylim(0, 300)  # excluding outliers with fare above 300
```

```python
num_data = data.drop(['Name', 'Embarked','Ticket', 'Cabin', 'Sex'], axis=1)

plt.figure(figsize=(10, 6))
corr = num_data.corr()
sns.heatmap(corr, annot=True, cmap='coolwarm', fmt='.2f', linewidths=0.5)
plt.title("Feature Correlation Heatmap")
plt.show()
```

```python
sns.pairplot(data[['Survived', 'Pclass', 'Age', 'Fare']], hue='Survived', palette='coolwarm')
plt.suptitle("Pairplot of Key Features by Survival", y=1.02)
plt.show()
```
