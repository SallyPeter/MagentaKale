```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, classification_report
```

```python
x = iris.drop(["target"], axis=1)  #(independent variable)
y = sklearn.datasets.load_iris().target

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.3, random_state=42)
```

```python
print(f"""x_train.shape: {x_train.shape}
x_test.shape: {x_test.shape} """)
```

```python
model = LogisticRegression()

model.fit(x_train, y_train)
```

```python
y_pred = model.predict(x_test)
```

```python
test_df = pd.DataFrame(x_test)
test_df["y_test"] = y_test
test_df["y_pred"] = y_pred

test_df
```

```python
accuracy_score(y_test, y_pred)
```

```python
print(classification_report(y_test, y_pred))
```

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
model_DT = DecisionTreeClassifier()

model_DT.fit(x_train, y_train)

y_pred_DT = model_DT.predict(x_test)
```

```python
test_df["y_pred_DT"] = y_pred_DT

test_df
```

```python
data.drop(["Customer_id","Interaction", "UID"],axis=1, inplace=True)
```

```python
data.info()
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
for each_col in data.select_dtypes(include='object').columns:
    data[each_col] = LabelEncoder().fit_transform(data[each_col])


data.info()
```

```python
x = data.drop(["ReAdmis"], axis=1)
y = data.ReAdmis

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.1, random_state=42)
```

```python
model = DecisionTreeClassifier(max_depth=10, criterion='entropy')

model.fit(x_train, y_train)

y_pred = model.predict(x_test)
```
